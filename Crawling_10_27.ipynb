### **뉴스 기사 제목 여러 개 가져오기**
- 페이지 수를 변경하면 뒷부분에 페이지 관련 url 부분 알 수 있음

```python
url1 = 'https://search.daum.net/search?w=news&DA=PGD&enc=utf8&cluster=y&cluster_page=1&q=%ED%8C%8C%EC%9D%B4%EC%8D%AC&p='

for i in range(1, 11):
    url = url1 + str(i)
    html = requests.get(url, headers = header).text
    soup = BeautifulSoup(html, 'html.parser')
    title = soup.find_all(class_ = 'tit_main fn_tit_u')
    for j in title:
        print(j.text, ':', j.attrs['href'])
    print(f'{i}페이지')
    print('-' * 100)
```

### **기사 제목 추출 함수 만들기(네이버)**

```python
from bs4 import BeautifulSoup
import requests

url1 = 'https://search.naver.com/search.naver?where=news&sm=tab_pge&query='
url3 = '&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=31&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all'
url4 = '&start='

def search_naver(url2, start_page, end_page):
    for i in range(start_page, end_page+1):
        url = url1 + url2 + url3 + url4 + str(i)
        html = requests.get(url).text
        soup = BeautifulSoup(html, 'html.parser')
        title = soup.find_all(class_ = 'news_tit')
        for j in title:
            print(j.text, ':', j.attrs['href'])
        print(f'{i}페이지입니다.')
        print('-' * 100)
```

```python
search_naver('다낭신', 1, 10)
```

### **데이터 프레임으로 만들기**

```python
import pandas as pd

def search_naver(word, start_page, end_page):
    li1 = []
    li2 = []
    
    for i in range(start_page, end_page+1):
        url = url1 + word + url3 + url4 + str(i)
        html = requests.get(url).text
        soup = BeautifulSoup(html, 'html.parser')
        title = soup.find_all(class_ = 'news_tit')
        for j in title:
            li1.append(j.text)
            li2.append(j.attrs['href'])
        news_df = pd.DataFrame({'title' : li1, 'url' : li2})
        news_df.to_csv(f'{word}.csv')
        return news_df
```

```python
search_naver('김연아', 1, 5)
```
